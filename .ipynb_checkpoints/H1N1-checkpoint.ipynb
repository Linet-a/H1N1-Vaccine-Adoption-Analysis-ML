 # H1N1 & Seasonal vaccine Adoption( Data Analysis & Modeling)

### Business Understanding

        

 Vaccination plays a pivotal role in infectuos disease management and the overal public health. While vaccination ensures a stable 'herd immunity many individuals remain unvaccinated due to various demographic, social and behavioural factors. This project aims to help the public health stakeholders to understand and predict vaccines uptake for H1N1 and seasonal flu vaccines. 
By analysing a dataset containing individual's demographic, opinion as well as health condition , we will come up with a model that predicts whether a person is likely to get vaccinated or not. This information is especially crucial to policy makers and healthcare providers in designing targeted awareness campaigns, efficient resource allocation in addressing vaccine uptake hessitancy in specific population group.

This informationn is very crual to the pharmaceutical companies and stakeholder dealing in the COVI-19 pandemics since most vaccines are still under development. By having H1N1 and seasonal flu model, it will be easier to adress problems that may arise with the vaccine uptake. The overal goal is to imporove vaccination rates and enhance public preparedness for future outbreaks.




### Data Understanding

The data consists of survey responses collected to analyse factors influencing vaccines (H1N1 and seasonal flu) uptake decisions. The data includes demographic information including age, gender, level of education and income level etc, health related issues e.g chronic conditions and health insurance  as well as behauvoral aspects including doctor's visit and personal risk perseption and lastly opinions on vaccines including trust in healthcare proffesionals and concerns about the vaccine's side effects.

Te datset is divided into:

 b. training labels - a target variable indication whether an individual received a vaccine or not

### Data Cleaning

```python
# checking for columns with missing values 
features_df.isna().sum()
```

```python
# employment_occupation, employment_industry, and health insurance columns have alot of missing values, 
#lets drop them
columns_to_drop = ['health_insurance', 'employment_industry', 'employment_occupation']
feature_df_cleaned = features_df.drop(columns=columns_to_drop)
```

```python
#fill the missing values in these columns with the mode since they are not much

cols_to_impute = ['h1n1_concern', 'h1n1_knowledge','behavioral_antiviral_meds', 'behavioral_avoidance',
                  'behavioral_face_mask','behavioral_wash_hands','behavioral_large_gatherings',
                  'behavioral_outside_home','behavioral_touch_face', 'education','income_poverty',
                  'doctor_recc_h1n1', 'doctor_recc_seasonal', 'chronic_med_condition',
                  'child_under_6_months', 'health_worker', 'household_adults', 'household_children']

for col in cols_to_impute:
    feature_df_cleaned[col].fillna(feature_df_cleaned[col].mode()[0
                                                                 ], inplace=True)
```

```python
#filling the missing opinion columns with the most frequent value

opinion_columns = [
    'opinion_h1n1_vacc_effective', 'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc',
    'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc'
]

# Option 1: Fill with Mode
for col in opinion_columns:
    feature_df_cleaned[col].fillna(feature_df_cleaned[col].mode()[0], inplace=True)

```

```python
print(feature_df_cleaned['marital_status'].value_counts())
print(feature_df_cleaned['rent_or_own'].value_counts())
print(feature_df_cleaned['employment_status'].value_counts())
print(feature_df_cleaned['income_poverty'].value_counts())
```

```python
# filling
feature_df_cleaned['rent_or_own'].fillna('Own', inplace=True)
feature_df_cleaned['marital_status'].fillna('Married', inplace=True)
feature_df_cleaned['employment_status'].fillna('Employed', inplace=True)                                            

                                           
```

# Building Models



## Logistic regresion - Baseline Model


```python
# import the necessary libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
from sklearn.metrics import roc_curve, auc
```

```python
# selecting the numeric columns for the base classification model
numeric_df = feature_label.select_dtypes(include=["int64", "float64"])
numeric_df


```

```python

#defining x and y
label_cols = ['h1n1_vaccine', 'seasonal_vaccine']
X = numeric_df.drop(columns=label_cols)
y = numeric_df[label_cols]

#splitting the data into test and train datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=31)

#scale the X_train and X_test

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#initialising logiststic reg class
logreg = LogisticRegression(penalty='l2',C=1.0, solver='liblinear', max_iter=1000) 

#using a multioutput classifier to fit the two models
multi_target_model = MultiOutputClassifier(logreg)
multi_target_model.fit(X_train_scaled, y_train)
```

```python


# predictions on train and test data
y_train_pred = multi_target_model.predict(X_train_scaled)
y_test_pred = multi_target_model.predict(X_test_scaled)

# probabilities for ROC AUC Score
y_train_proba = multi_target_model.predict_proba(X_train_scaled)
y_test_proba = multi_target_model.predict_proba(X_test_scaled)

# accuracy for both targets
train_acc_h1n1 = accuracy_score(y_train['h1n1_vaccine'], y_train_pred[:, 0])
test_acc_h1n1 = accuracy_score(y_test['h1n1_vaccine'], y_test_pred[:, 0])

train_acc_seasonal = accuracy_score(y_train['seasonal_vaccine'], y_train_pred[:, 1])
test_acc_seasonal = accuracy_score(y_test['seasonal_vaccine'], y_test_pred[:, 1])

# Calculating ROC AUC Score
train_auc_h1n1 = roc_auc_score(y_train['h1n1_vaccine'], y_train_proba[0][:, 1])
test_auc_h1n1 = roc_auc_score(y_test['h1n1_vaccine'], y_test_proba[0][:, 1])

train_auc_seasonal = roc_auc_score(y_train['seasonal_vaccine'], y_train_proba[1][:, 1])
test_auc_seasonal = roc_auc_score(y_test['seasonal_vaccine'], y_test_proba[1][:, 1])

# printing  accuracy results
print(f"H1N1 Vaccine - Train Accuracy: {train_acc_h1n1:.4f}, Test Accuracy: {test_acc_h1n1:.4f}")
print(f"H1N1 Vaccine - Train AUC: {train_auc_h1n1:.4f}, Test AUC: {test_auc_h1n1:.4f}")

print(f"Seasonal Vaccine - Train Accuracy: {train_acc_seasonal:.4f}, Test Accuracy: {test_acc_seasonal:.4f}")
print(f"Seasonal Vaccine - Train AUC: {train_auc_seasonal:.4f}, Test AUC: {test_auc_seasonal:.4f}")

# printing classification reports
print("\nH1N1 Vaccine - Classification Report (Test Set):")
print(classification_report(y_test['h1n1_vaccine'], y_test_pred[:, 0]))

print("\nSeasonal Vaccine - Classification Report (Test Set):")
print(classification_report(y_test['seasonal_vaccine'], y_test_pred[:, 1]))
```

### Visualizing the ROC AUC Curve

```python

# a function to plot ROC Curve
def plot_roc_curve(y_true, y_proba, label, dataset_type):
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{label} ({dataset_type}) AUC: {roc_auc:.3f}')
    return roc_auc


plt.figure(figsize=(10, 6))

# ROC for H1N1 Vaccine
auc_train_h1n1 = plot_roc_curve(y_train['h1n1_vaccine'], y_train_proba[0][:, 1], 'H1N1 Vaccine', 'Train')
auc_test_h1n1 = plot_roc_curve(y_test['h1n1_vaccine'], y_test_proba[0][:, 1], 'H1N1 Vaccine', 'Test')

# ROC for Seasonal Vaccine
auc_train_seasonal = plot_roc_curve(y_train['seasonal_vaccine'], y_train_proba[1][:, 1], 'Seasonal Vaccine', 'Train')
auc_test_seasonal = plot_roc_curve(y_test['seasonal_vaccine'], y_test_proba[1][:, 1], 'Seasonal Vaccine', 'Test')

# diagonal line (random classifier)
plt.plot([0, 1], [0, 1], 'k--', label="Random Classifier (AUC=0.5)")


plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve for H1N1 & Seasonal Vaccine Predictions")
plt.legend()
plt.show()
```

## Model Limitation


This model struggles with class imbalance as very few people were vaccinated against h1n1 virus. Therefore,there is a need for a more complex approach that handles class imbalace especially in h1n1-vaccine target

```python

# Defining the parameter grid
param_grid = {
    "max_depth": [5, 10, 15, 20, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 5]
}
```
